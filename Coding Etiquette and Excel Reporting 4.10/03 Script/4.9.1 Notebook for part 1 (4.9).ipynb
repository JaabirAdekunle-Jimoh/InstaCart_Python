# 5 Complete the fundamental data quality and consistency checks

```python
# Check info about columns with missing values.
df_custs.isnull().sum()

## There are 11259 missing 'First Names'
```

```python
# Create a subset of the dataframe containing only the missing values in question
df_nan = df_custs[df_custs['First_Name'].isnull() ==True]
```

```python
df_nan.head()
```

```python
# Rows with missing values.
df_nan.shape
```

```python
# Deal with the missing data.
## Since the column with the missing data is the "First_Name" which is a 'string'. 
## We solve this by filtering out those not missing into a subset dataframe for our analysis.
df_custs_clean = df_custs[df_custs['First_Name'].isnull() == False]
```

```python
# Rows without missing values.
df_custs_clean.shape

## 11259 + 194950 = 206209 (this shows the subset is correct)
```

```python
# check for and address duplicates
df_dups = df_custs_clean[df_custs_clean.duplicated()]
```

```python
df_dups.head()
```

```python
## ## This means there are no duplicates in the df_custs_clean dataframe 
```

```python
# Check for data consistency again

for col in df_custs_clean.columns.tolist():
  weird = (df_custs_clean[[col]].applymap(type) != df_custs_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_custs_clean[weird]) > 0:
    print (col)
  else:
    print ('No mixed dtypes', col)
```

```python
# convert any mixed-type data
## 'User_id' needs to be converted from int64 to str since it cannot be used in calculations.
## Changing data type of 'User_id' from int to str
df_custs_clean['User_id'] = df_custs_clean['User_id'].astype('str')
```

```python
df_custs_clean.info()
```

```python
df_custs_clean['User_id'].dtype
```

```python
# Export the cleaned Customer Dataframe as pickle
df_custs_clean.to_pickle(os.path.join(path,'02 Data','Prepared Data','custs_checked.pkl'))
```
