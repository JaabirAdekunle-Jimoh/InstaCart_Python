# Ex 4.7.2

1. Create the Busiest_days and Slowest_days column

```python
# # Printing the frequency of the “orders_day_of_the_week” column
df_ords_prods_merge['order_dow'].value_counts(dropna = False)
```

Observation: The two busiest days are 0 (Saturday) and 1 (Sunday) while the two slowest days are 3 (Tuesday) and 4 ( Wednesday). 

2. Create new columns

```python
# Assign the corresponding string value to the day frequencies.

days = []

for day in df_ords_prods_merge["order_dow"]:
   if day == 0 or day == 1:
    days.append('Busiest days')
   elif day == 4 or day == 3:
    days.append('Slowest days')
   else:
    days.append('Regular days')
```

```python
days
```

```python
# create the 'Busiest_days' column
df_ords_prods_merge['Busiest_days'] = days

# get the frequency of the values in the 'busiest_days' column
df_ords_prods_merge['Busiest_days'].value_counts(dropna = False)
```

3. Check the values of this new column for accuracy.

```python
# check the output of the 'order_dow' and the 'busiest_days' column for accuracy
print('Output of the order_dow and the Busiest_days column:\n', df_ords_prods_merge[['order_dow', 'Busiest_days']].head(10))
```

4. Create the 'busiest_period_of_day' column

```python
# get the frequency of the values in the 'order_hour_of_day' column
print('Frequency of values in the order_hour_of_day column:\n', df_ords_prods_merge['order_hour_of_day'].value_counts())
```

Observations: The 24 hours of the day can be divided into three parts;

'Most orders' (hour 10, 11, 14, 15, 13, 12, 16, 9 - from 9am to 4pm),

'Average orders' (hour 17, 8, 18, 19, 20, 7, 21, 22 - from 7am to 8am and from 5pm to 10pm), or 

'Fewest orders' (hour 23, 6, 0, 1, 5, 2, 4, 3 - from 11pm to 6am).

```python
# create a for-loop that append to the 'hours' list the busyness of the hours based on values in the 'order_hour_of_day' column
hours = []

for hour in df_ords_prods_merge['order_hour_of_day']:
  if hour in [10, 11, 14, 15, 13, 12, 16, 9]:
    hours.append('Most orders')
  elif hour in [23, 6, 0, 1, 5, 2, 4, 3]:
    hours.append('Fewest orders')
  else:
    hours.append('Average orders')
```

```python
# create the 'busiest_period_of_day' column
df_ords_prods_merge['busiest_period_of_day'] = hours
```

```python
# check the output of the 'order_hour_of_day' and the 'busiest_period_of_day' column for accuracy
df_ords_prods_merge[['order_hour_of_day', 'busiest_period_of_day']].head(10)
```

5. Print the frequency for this new column.

```python
# get the frequency of the values in the 'busiest_period_of_day' column
print('Frequency of values in the busiest_period_of_day column:\n', df_ords_prods_merge['busiest_period_of_day'].value_counts())
```
