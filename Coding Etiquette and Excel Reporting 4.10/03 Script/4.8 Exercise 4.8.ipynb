##4 Create a flag that assigns a “loyalty” label to existing customers using the transform() and loc() functions.
To do this, maximum order ("max_order") column must be created.

```python
# create your “max_order” column. In this case, all three steps can be included in a single line of code:
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the created column
df_ords_prods_merge.head(20)
```

```python
# Using the loc function
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print the frequency of your new “loyalty_flag” column using the value_counts() function:
df_ords_prods_merge ['loyalty_flag'].value_counts(dropna = False)
```

##5 Trying to determine whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Performing Multiple Aggregations
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})

## The result shows that 'Loyal_customer' purchased the CHEAPEST products,
##'Regular_customer' purchased MORE EXPENSIVE products while the 'New_customer' purchased the MOST EXPENSIVE products. 
```

##6 Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# create your “avg_price” column. In this case, all three steps can be included in a single line of code:
df_ords_prods_merge['avg_price'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the created column
df_ords_prods_merge.head()
```

```python
# Create a spending flag for each user based on the average price across all their orders using the following criteria:
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'high spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Print the frequency of your new “avg_price” column using the value_counts() function:
df_ords_prods_merge ['spending_flag'].value_counts(dropna = False)
```

##7 Determine frequent versus non-frequent customers. 

##Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# create your “median_dspo” column. In this case, all three steps can be included in a single line of code:
df_ords_prods_merge['median_dspo'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the created “median_dspo” column
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_dspo'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_dspo'] > 10) & (df_ords_prods_merge['median_dspo'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_dspo'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Print the frequency of your new “avg_price” column using the value_counts() function:
df_ords_prods_merge ['order_frequency_flag'].value_counts(dropna = False)
```
